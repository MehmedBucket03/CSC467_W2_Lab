```python
# Mean filter Lena.png

import matplotlib.pyplot as plt

import urllib.request # Importing the urllib.request module

url = ('https://github.com/MehmedBucket03/CSC467_W2_Lab/blob/main/Lena.png?raw=true')
resp = urllib.request.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype = "uint8") # Corrected 'rep' to 'resp'
image = cv2.imdecode(image, cv2.IMREAD_GRAYSCALE)
mean_filter = cv2.blur(image, (5,5))


plt.figure(figsize = (10,10))
plt.subplot(1,2,1), plt.imshow(image, cmap = "gray"), plt.title("Original")
plt.subplot(1,2,2), plt.imshow(mean_filter, cmap = "gray"), plt.title("Mean Filter")
plt.show()
```

```python
# Mean filter Baboon.png

import matplotlib.pyplot as plt

import urllib.request # Importing the urllib.request module

url = ('https://github.com/MehmedBucket03/CSC467_W2_Lab/blob/main/baboon.png?raw=true')
resp = urllib.request.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype = "uint8") # Corrected 'rep' to 'resp'
image = cv2.imdecode(image, cv2.IMREAD_GRAYSCALE)
mean_filter = cv2.blur(image, (5,5))


plt.figure(figsize = (10,10))
plt.subplot(1,2,1), plt.imshow(image, cmap = "gray"), plt.title("Original")
plt.subplot(1,2,2), plt.imshow(mean_filter, cmap = "gray"), plt.title("Mean Filter")
plt.show()
```

```python
# Sharpening tool Lena.png

sharpening_kernel = np.array([[0,-1,0],
                             [-1,5,-1],
                             [0,-1,0]])
sharpened_image = cv2.filter2D(image, -1, sharpening_kernel)

plt.figure(figsize = (10,10))
plt.subplot(1,2,1), plt.imshow(image, cmap = "gray"), plt.title("Original")
plt.subplot(1,2,2), plt.imshow(sharpened_image, cmap = "gray"), plt.title("Sharpened")
plt.show()
```

```python
# Sharpening tool Baboon.png

sharpening_kernel = np.array([[0,-1,0],
                             [-1,5,-1],
                             [0,-1,0]])
sharpened_image = cv2.filter2D(image, -1, sharpening_kernel)

plt.figure(figsize = (10,10))
plt.subplot(1,2,1), plt.imshow(image, cmap = "gray"), plt.title("Original")
plt.subplot(1,2,2), plt.imshow(sharpened_image, cmap = "gray"), plt.title("Sharpened")
plt.show()
```

```python
# edge detection Lena.png

url = "https://github.com/MehmedBucket03/CSC467_W2_Lab/blob/main/Lena.png?raw=true"
resp = urllib.request.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype="uint8")
image = cv2.imdecode(image, cv2.IMREAD_GRAYSCALE)

sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize = 5)
sobel_combined = np.sqrt(sobelx**2 + sobely**2)

sobel_combined = cv2.normalize(sobel_combined, None, 0, 255, cv2.NORM_MINMAX)
sobel_combined = np.uint8(sobel_combined)
edges = cv2.Canny(image, 100, 200)

plt.figure(figsize = (10,10))
plt.subplot(1,3,1), plt.imshow(image, cmap = "gray"), plt.title("Original")
plt.subplot(1,3,2), plt.imshow(sobel_combined, cmap = "gray"), plt.title("Sobel")
plt.subplot(1,3,3), plt.imshow(edges, cmap = "gray"), plt.title("Canny")
plt.show()
```

```python
# edge detection Baboon.png

url = "https://github.com/MehmedBucket03/CSC467_W2_Lab/blob/main/baboon.png?raw=true"
resp = urllib.request.urlopen(url)
image = np.asarray(bytearray(resp.read()), dtype="uint8")
image = cv2.imdecode(image, cv2.IMREAD_GRAYSCALE)

sobelx = cv2.Sobel(image, cv2.CV_64F, 1, 0, ksize = 5)
sobely = cv2.Sobel(image, cv2.CV_64F, 0, 1, ksize = 5)
sobel_combined = np.sqrt(sobelx**2 + sobely**2)

sobel_combined = cv2.normalize(sobel_combined, None, 0, 255, cv2.NORM_MINMAX)
sobel_combined = np.uint8(sobel_combined)
edges = cv2.Canny(image, 100, 200)

plt.figure(figsize = (10,10))
plt.subplot(1,3,1), plt.imshow(image, cmap = "gray"), plt.title("Original")
plt.subplot(1,3,2), plt.imshow(sobel_combined, cmap = "gray"), plt.title("Sobel")
plt.subplot(1,3,3), plt.imshow(edges, cmap = "gray"), plt.title("Canny")
plt.show()
```
